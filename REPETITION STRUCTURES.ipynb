**while loop**
- enter while loop of condition is true, after completing loop body, re-evaluates condition and if true, execute body again 
1. condition tested for true or false value
2. statements repeated as long as condition is true
- then line will loop

while condition
  - statement1
  - statement2
  - statement3

```python
balance = float(input("How much money do you have?"))
while balance>0:
    price=float(input("enter item price"))
    if price<=balance:
        print("Item Purchased!")
        balance-=price
    else:
        print("Insuffient Funds")
    print("current balance:", balance)
```

```python
#condition is false
#can make infinite loop by commenting out second "guess" input
secretNumber=9
guess = int(input("Try to guess secret number (1-10)!"))
attempts=1

while guess!=secretNumber:
    print(attempts, "this is not it, try again")
    guess = int(input("Try to guess secret number (1-10)!"))
    attempts+=1
print("That's it! Congrats!!")
```

```python
import random

secretNumber=random.randint(1,10)
guess = int(input("Try to guess secret number (1-10)! [-1 to exit]"))
attempts=1

while guess!=secretNumber and guess!=-1:
    print(attempts, "this is not it, try again")
    guess = int(input("Try to guess secret number (1-10)! [-1 to exit]"))
    attempts+=1
if guess==secretNumber:
    print("thats it! congrats)")
print("That's it! Congrats!!")
```

```python
balance = float(input("How much money do you have?"))
while balance>0:
    price=float(input("enter item price"))
    while price<=0:
        print("invalid price for item")
        price=float(input("enter item price"))

    
    if price<=balance:
        print("Item Purchased!")
        balance-=price
    else:
        print("Insuffient Funds")
    print("current balance:", balance) 
```

```python
sum=0
count=0
for number in [12,24,324,2,5,]:
    sum+=number
    count+=1
    print("processing", number)
print("the sum of those numbers is", sum)
print("there are", count, "numbers in that list")
```

### nested loops

- loop within a loop
- no limit to amount
- inner most loop runs more than outer loops

```python
for course in ["CS0010", "CS0141", "CS0011", "CS0001", "CS1550"]:
    print("Preparing to study for:", course)

    for step in ["read the book", "review notes", "solve problems", "study flashcards", "take the exam"]:
        if course == "CS0001" and step=="read the book":
            continue
        elif (course == "CS0011" or course=="CS0010" and step == "review notes"):
            continue
        elif course in ("CS0010","CS0011") and step=="review notes":
            continue
        print(step)
    print("done with", course)
    print("-"*30)
```

```python
import time
periodOfDay="AM"

while True:
    for hour in range(12):
        for minute in range(60):
            for second in range(60):
                hour=format(hour, "0>2")
                minute=format(minute, "0>2")
                second=format(second, "0>2")
                print(hour, minute, second+periodOfDay, sep=":")
                #time.sleep(.05)
    if periodOfDay=="AM":
        periodOfDay="PM"
    else:
        periodOfDay=="AM"
```
